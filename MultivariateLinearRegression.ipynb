```python
import numpy as np
import pandas as pd
import scipy as sp
import statsmodels.api as sm
import datetime as dt
import pandas_datareader.data as web

start = dt.datetime(2011,12,31)
end = dt.datetime(2016,12,31)
ticker='IBM'

p =web.DataReader(ticker, 'yahoo', start=start, end=end)
p.head()
```

```python
monthly = p.asfreq('BM')
monthly.head()
```

```python
monthly['ret'] = ((monthly['Adj Close'] / monthly['Adj Close'].shift(1)) - 1)*100
monthly2 = monthly.dropna()
monthly2.head()
```

```python
factors = web.DataReader("F-F_Research_Data_5_Factors_2x3", "famafrench",start='2012-01-01', end ='2016-12-31')[0]
factors = factors.reset_index(drop=True)
factors['Date'] = pd.date_range(start='2012-01-01', end ='2016-12-31',  freq='BM')
factors = factors.set_index(['Date'])
factors.head()
```

```python
final = pd.merge(monthly2, factors, left_index = True, right_index = True)
final.head()
```

```python
final['ret-RF'] = final['ret'] - final['RF'] 
```

```python
x2 = final[['Mkt-RF','SMB','HML']]
y2 = final['ret-RF']
x2 = sm.add_constant(x2)
result2 = sm.OLS(y2,x2).fit()
print(result2.summary())
```
